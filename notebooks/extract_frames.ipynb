```python
import cv2
vidcap = cv2.VideoCapture('/home/thyde/avatar-stuff/eps/season1/s01e09.mkv')
success,image = vidcap.read()
count = 0
while success:
    cv2.imwrite("/home/thyde/avatar-stuff/frames/s01e09/raw_frames/frame%d.jpg" % count, image)     # save frame as JPEG file      
    success,image = vidcap.read()
    print('Read a new frame: ', success)
    count += 1
```
